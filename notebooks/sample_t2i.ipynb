```python
import os
from omegaconf import OmegaConf
import hydra
import torch
os.environ['HYDRA_FULL_ERROR'] = '1'
os.environ["CUDA_VISIBLE_DEVICES"] = "5"
import sys
sys.path.append('..')
import diffusion

cfg = OmegaConf.load('../configs/model/rf_dit_t2i.yaml')
model = hydra.utils.instantiate(cfg)
model = model.cuda()

model
```

```python
ckpt_pth = '/export/home/ra49veb/dev/min-diffusion/diffusion/t2i.pt'
model.load_state_dict({ k: v for k, v in torch.load(ckpt_pth, map_location='cuda').items() if not 'ae' in k }, strict=False)
model.eval()
```

```python
bs = 8
z = torch.randn(bs, 4, 32, 32).cuda()
caption = ['a photo of a tree'] * bs
```

```python
model.cfg_scale = 5

with torch.no_grad():
    res = model.sample(z=z, txt=caption)
```

```python
from torchvision.transforms.functional import to_pil_image
img = res.clip(-1, 1).float().cpu() * 0.5 + 0.5
from einops import rearrange
img = rearrange(img, 'b c h w -> c h (b w)')
to_pil_image(img)
```
